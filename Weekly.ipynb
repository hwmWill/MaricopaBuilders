```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Get spreadsheet links for 2023 and 2022
r = requests.get('https://www.maricopa.gov/Archive.aspx?AMID=128')
soup = BeautifulSoup(r.content)
urls = soup.find('div', {'id':'modulecontent'}).find_all('a', href=True)[1:83]
urls = ['https://www.maricopa.gov/'+x['href'] for x in urls]
```

```python
df = pd.DataFrame()

for url in urls:
    print(url)
    r = requests.get(url)
    df = pd.concat([df, pd.read_excel(r.content)], ignore_index=True)
```

```python
newHousing = df.loc[(df.DESCRIPTION=='Residential')&(df.CATEGORY.isin(['New', 'New with Basement', 'Production']))].reset_index(drop=True)
newHousing.info()
```
